Q2 Write a program in python to assign subsequent rows to matrix first row elements.

```python
import numpy as np

# Create sample matrix
matrix = np.random.randint(1,10,(4,5))

print("Original Matrix:\n", matrix)

# Assign all subsequent rows to first row
matrix[1:, :] = matrix[0, :]

print("\nModified Matrix:\n", matrix)
```

Q3 - Write a program in python to group similar elements into matrix.

```python
import numpy as np

data = np.array([1,2,1,3,2,1,4,3,6])

unique_vals = np.unique(data)

groups = [data[data == val] for val in unique_vals]

print("Grouped Elements:")
for row in groups:
    print(row)
```

Q4 - Write a program in python to add and subtract matrices in python.

```python
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

print(f"Enter {rows} rows for Matrix A, each with {cols} elements separated "
      "by spaces:")
A = []
for i in range(rows):
    while True:
        try:
            row_input = list(map(int, input().split()))
            if len(row_input) == cols:
                A.append(row_input)
                break
            else:
                print(f"Enter {cols} elements per row. Try again.")
        except ValueError:
            print("Invalid input. Integers only. Try again.")

print(f"Enter {rows} rows for Matrix B, each with {cols} elements separated "
      "by spaces:")
B = []
for i in range(rows):
    while True:
        try:
            row_input = list(map(int, input().split()))
            if len(row_input) == cols:
                B.append(row_input)
                break
            else:
                print(f"Enter {cols} elements per row. Try again.")
        except ValueError:
            print("Invalid input. Integers only. Try again.")

add_result = [[0]*cols for _ in range(rows)]
sub_result = [[0]*cols for _ in range(rows)]

for i in range(rows):
    for j in range(cols):
        add_result[i][j] = A[i][j] + B[i][j]
        sub_result[i][j] = A[i][j] - B[i][j]

print("\nAddition Result:")
for row in add_result:
    print(row)

print("\nSubtraction Result:")
for row in sub_result:
    print(row)
```

Q5 - Write a program in python to create a n*n square matrix, where all the sub-matrix has the sum of opposite corner elements as even.

```python
def create_matrix(n):
    matrix = [[(i + j) for j in range(n)] for i in range(n)]
    return matrix

# Input
n = int(input("Enter value of n: "))

# Generate matrix
matrix = create_matrix(n)

# Print matrix
print("\nGenerated Matrix:")
for row in matrix:
    print(row)
```

Q6 - Write a program in python to perform row-wise element addition in tuple matrix.

```python
matrix = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9)
)

# row-wise sum
row_sums = tuple(sum(row) for row in matrix)

print("Row-wise sums:", row_sums)
```
